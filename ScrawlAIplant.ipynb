```python
#多线程下载数据
import urllib.request
import socket
from urllib import error
from queue import Queue
from threading import Thread
import os
import pickle
class DownloadWorker(Thread):  #定义一个类，继承自thread类，重写其run函数
    def __init__(self, queue):
        Thread.__init__(self)
        self.queue = queue     #标准的多线程实现方法都使用了queue
    def run(self):
        while True:
            link,basepath,pro_name,class_name,file_name = self.queue.get() #从队列中获取一组网址以及对应的保存位置
            outpath = os.path.join(basepath,pro_name,class_name)
            if not os.path.exists(outpath):
                os.makedirs(outpath)
            outfname = os.path.join(outpath,file_name)
            try: #使用try except方法进行各种异常处理
                response = urllib.request.urlopen(link, timeout=5)
                data = response.read()
                with open(outfname, 'wb') as imfid:
                    imfid.write(data)
                print(link)
            except error.HTTPError as err:
                print('HTTPerror, code: %s' % err.code)
            except error.URLError as err:
                print('URLerror, reason: %s' % err.reason)
            except socket.timeout:
                print('Time Out!')
            except:
                print('Unkown Error!')
            self.queue.task_done() #标记队列中的一个元素已经被处理
def main():
    fd = open(r"F:\AI_iplant\Rest_Paramaters.txt","rb")
    parameters = pickle.load(fd)
    fd.close()
    queue = Queue() #定义队列
    for x in range(12): #开启12个线程
        worker = DownloadWorker(queue)
        worker.daemon = True
        worker.start()
    for lineData in parameters: #向队列中放入数据
        link = lineData[0]
        inpath = lineData[1]
        projname = lineData[2]
        classname = lineData[3]
        filename = lineData[4]
        if not os.path.exists(os.path.join(inpath,projname,classname,filename)):
            queue.put((link,inpath,projname,classname,filename))
    queue.join() #等待队列中的数据被处理完毕
if __name__ == '__main__':
    main()
```
