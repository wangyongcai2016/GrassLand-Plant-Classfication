```python
#Circular barplot with groups   绘制图像数量前10的科
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plt.rcParams["font.sans-serif"] = ["SimHei"]    #用来正常显示中文标签
plt.rcParams["axes.unicode_minus"] = False    #用来正常显示中文标签
# Ensures reproducibility of random numbers
rng = np.random.default_rng(123)
df = pd.read_excel(r"E:\论文\草地植物分类\model\每科前10_top5.xlsx")
def get_label_rotation(angle, offset):
    # Rotation must be specified in degrees :(
    rotation = np.rad2deg(angle + offset)
    if angle <= np.pi:
        alignment = "right"
        rotation = rotation + 180
    else:
        alignment = "left"
    return rotation, alignment
def add_labels(angles, values, labels, offset, ax):

    # This is the space between the end of the bar and the label
    padding = 4

    # Iterate over angles, values, and labels, to add all of them.
    for angle, value, label, in zip(angles, values, labels):
        angle = angle

        # Obtain text rotation and alignment
        rotation, alignment = get_label_rotation(angle, offset)

        # And finally add the text
        ax.text(
            x=angle,
            y=value + padding,
            s=label,
            ha=alignment,
            va="center",
            rotation=rotation,
            rotation_mode="anchor"
        )
# All this part is like the code above
VALUES = df["value"].values
LABELS = df["Family"].values
GROUP = df["Model"].values

PAD = 2
ANGLES_N = len(VALUES) + PAD * len(np.unique(GROUP))
ANGLES = np.linspace(0, 2 * np.pi, num=ANGLES_N, endpoint=False)
WIDTH = (2 * np.pi) / len(ANGLES)
OFFSET = np.pi / 2
offset = 0
IDXS = []
GROUPS_SIZE = [10, 10, 10, 10]
for size in GROUPS_SIZE:
    IDXS += list(range(offset + PAD, offset + size + PAD))
    offset += size + PAD

fig, ax = plt.subplots(figsize=(20, 10), subplot_kw={"projection": "polar"})
ax.set_theta_offset(OFFSET)
ax.set_ylim(-80, 100)
ax.set_frame_on(False)
ax.xaxis.grid(False)
ax.yaxis.grid(False)
ax.set_xticks([])
ax.set_yticks([])

GROUPS_SIZE = [10, 10, 10, 10]
COLORS = [f"C{i}" for i, size in enumerate(GROUPS_SIZE) for _ in range(size)]

ax.bar(
    ANGLES[IDXS], VALUES, width=WIDTH, color=COLORS,
    edgecolor="white", linewidth=2
)

add_labels(ANGLES[IDXS], VALUES, LABELS, OFFSET, ax)

# Extra customization below here --------------------

# This iterates over the sizes of the groups adding reference
# lines and annotations.

offset = 0
model_name_angle = [60,-45,45,-45]
for group,ang,size in zip(["Eva02", "MobileVitV2", "Resnetrs", "MobileNetV3"],model_name_angle,GROUPS_SIZE):      #显示分组标签
    # Add line below bars
    x1 = np.linspace(ANGLES[offset + PAD], ANGLES[offset + size + PAD - 1], num=40)
    ax.plot(x1, [-5] * 40, color="#333333")

    # Add text to indicate group
    ax.text(
        np.mean(x1), -20, group, color="#333333", fontsize=14,
        fontweight="bold", ha="center", va="center",
        rotation=ang,
        rotation_mode="anchor"
    )
    # Add reference lines at 20, 40, 60, and 80
    x2 = np.linspace(ANGLES[offset], ANGLES[offset + PAD - 1], num=40)
    ax.plot(x2, [0] * 40, color="#bebebe", lw=0.8)
    ax.plot(x2, [20] * 40, color="#bebebe", lw=0.8)
    ax.plot(x2, [40] * 40, color="#bebebe", lw=0.8)
    ax.plot(x2, [60] * 40, color="#bebebe", lw=0.8)
    ax.plot(x2, [80] * 40, color="#bebebe", lw=0.8)

    ax.text(0 * np.pi / 2 + 0.05, 0 + PAD, "0", ha="center", size=12,color="deepskyblue")
    ax.text(0 * np.pi / 2 + 0.05, 20 + PAD, "20%", ha="center", size=12,color="deepskyblue")
    ax.text(0 * np.pi / 2 + 0.05, 40 + PAD, "40%", ha="center", size=12,color="deepskyblue")
    ax.text(0 * np.pi / 2 + 0.05, 60 + PAD, "60%", ha="center", size=12,color="deepskyblue")
    ax.text(0 * np.pi / 2 + 0.05, 80 + PAD, "80%", ha="center", size=12,color="deepskyblue")

    ax.text(1 * np.pi / 1.9, 0 + PAD, "0", ha="center", size=12,color="darkorange")
    ax.text(1 * np.pi / 1.9, 20 + PAD, "20%", ha="center", size=12,color="darkorange")
    ax.text(1 * np.pi / 1.9, 40 + PAD, "40%", ha="center", size=12,color="darkorange")
    ax.text(1 * np.pi / 1.9, 60 + PAD, "60%", ha="center", size=12,color="darkorange")
    ax.text(1 * np.pi / 1.9, 80 + PAD, "80%", ha="center", size=12,color="darkorange")

    ax.text(2 * np.pi / 2 + 0.06, 0 + PAD, "0", ha="center", size=12,color="lightgreen")
    ax.text(2 * np.pi / 2 + 0.06, 20 + PAD, "20%", ha="center", size=12,color="lightgreen")
    ax.text(2 * np.pi / 2 + 0.06, 40 + PAD, "40%", ha="center", size=12,color="lightgreen")
    ax.text(2 * np.pi / 2 + 0.06, 60 + PAD, "60%", ha="center", size=12,color="lightgreen")
    ax.text(2 * np.pi / 2 + 0.06, 80 + PAD, "80%", ha="center", size=12,color="lightgreen")

    ax.text(3 * np.pi / 2 + 0.05, 0 + PAD, "0", ha="center", size=12,color="lightcoral")
    ax.text(3 * np.pi / 2 + 0.05, 20 + PAD, "20%", ha="center", size=12,color="lightcoral")
    ax.text(3 * np.pi / 2 + 0.05, 40 + PAD, "40%", ha="center", size=12,color="lightcoral")
    ax.text(3 * np.pi / 2 + 0.05, 60 + PAD, "60%", ha="center", size=12,color="lightcoral")
    ax.text(3 * np.pi / 2 + 0.05, 80 + PAD, "80%", ha="center", size=12,color="lightcoral")

    offset += size + PAD
plt.savefig(r"E:\论文\草地植物分类\model\Family_前10_top5.jpg",dpi=300)
```

```python
#Circular barplot with groups  绘制图像数量前20科
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plt.rcParams["font.sans-serif"] = ["SimHei"]    #用来正常显示中文标签
plt.rcParams["axes.unicode_minus"] = False    #用来正常显示中文标签
# Ensures reproducibility of random numbers
rng = np.random.default_rng(123)
df = pd.read_excel(r"E:\论文\草地植物分类\model\每科前20 - new.xlsx",sheet_name=0)
def get_label_rotation(angle, offset):
    # Rotation must be specified in degrees :(
    rotation = np.rad2deg(angle + offset)
    if angle <= np.pi:
        alignment = "right"
        rotation = rotation + 180
    else:
        alignment = "left"
    return rotation, alignment
def add_labels(angles, values, labels, offset, ax):

    # This is the space between the end of the bar and the label
    padding = 4

    # Iterate over angles, values, and labels, to add all of them.
    for angle, value, label, in zip(angles, values, labels):
        angle = angle

        # Obtain text rotation and alignment
        rotation, alignment = get_label_rotation(angle, offset)

        # And finally add the text
        ax.text(
            x=angle,
            y=value + padding,
            s=label,
            ha=alignment,
            va="center",
            rotation=rotation,
            rotation_mode="anchor"
        )
# All this part is like the code above
VALUES = df["value"].values
LABELS = df["Family"].values
GROUP = df["Model"].values

PAD = 2
ANGLES_N = len(VALUES) + PAD * len(np.unique(GROUP))
ANGLES = np.linspace(0, 2 * np.pi, num=ANGLES_N, endpoint=False)
WIDTH = (2 * np.pi) / len(ANGLES)
OFFSET = np.pi / 2
offset = 0
IDXS = []
GROUPS_SIZE = [20, 20, 20, 20]
for size in GROUPS_SIZE:
    IDXS += list(range(offset + PAD, offset + size + PAD))
    offset += size + PAD

fig, ax = plt.subplots(figsize=(10, 10), subplot_kw={"projection": "polar"})
ax.set_theta_offset(OFFSET)
ax.set_ylim(-80, 100)
ax.set_frame_on(False)
ax.xaxis.grid(False)
ax.yaxis.grid(False)
ax.set_xticks([])
ax.set_yticks([])

GROUPS_SIZE = [20, 20, 20, 20]
COLORS = [f"C{i}" for i, size in enumerate(GROUPS_SIZE) for _ in range(size)]

ax.bar(
    ANGLES[IDXS], VALUES, width=WIDTH, color=COLORS,
    edgecolor="white", linewidth=2
)

add_labels(ANGLES[IDXS], VALUES, LABELS, OFFSET, ax)

# Extra customization below here --------------------

# This iterates over the sizes of the groups adding reference
# lines and annotations.

offset = 0
model_name_angle = [60,-45,45,-45]
for group,ang,size in zip(["Eva02", "MobileVitV2", "Resnetrs", "MobileNetV3"],model_name_angle,GROUPS_SIZE):      #显示分组标签
    # Add line below bars
    x1 = np.linspace(ANGLES[offset + PAD], ANGLES[offset + size + PAD - 1], num=80)
    ax.plot(x1, [-5] * 80, color="#333333")

    # Add text to indicate group
    ax.text(
        np.mean(x1), -20, group, color="#333333", fontsize=14,
        fontweight="bold", ha="center", va="center",
        rotation=ang,
        rotation_mode="anchor"
    )
    # Add reference lines at 20, 40, 60, and 80
    x2 = np.linspace(ANGLES[offset], ANGLES[offset + PAD - 1], num=80)
    ax.plot(x2, [0] * 80, color="#bebebe", lw=0.8)
    ax.plot(x2, [20] * 80, color="#bebebe", lw=0.8)
    ax.plot(x2, [40] * 80, color="#bebebe", lw=0.8)
    ax.plot(x2, [60] * 80, color="#bebebe", lw=0.8)
    ax.plot(x2, [80] * 80, color="#bebebe", lw=0.8)
    ax.plot(x2, [100] * 80, color="#bebebe", lw=0.8)

    ax.text(0 * np.pi / 2 + 0.03, 0 + PAD, "0%", ha="center", size=12,color="deepskyblue")
    ax.text(0 * np.pi / 2 + 0.03, 20 + PAD, "20%", ha="center", size=12,color="deepskyblue")
    ax.text(0 * np.pi / 2 + 0.03, 40 + PAD, "40%", ha="center", size=12,color="deepskyblue")
    ax.text(0 * np.pi / 2 + 0.03, 60 + PAD, "60%", ha="center", size=12,color="deepskyblue")
    ax.text(0 * np.pi / 2 + 0.03, 80 + PAD, "80%", ha="center", size=12,color="deepskyblue")
    ax.text(0 * np.pi / 2 + 0.03, 100 + PAD, "100%", ha="center", size=12,color="deepskyblue")

    ax.text(1 * np.pi / 2 + 0.05, 0 + PAD, "0%", ha="center", size=12,color="darkorange")
    ax.text(1 * np.pi / 2 + 0.05, 20 + PAD, "20%", ha="center", size=12,color="darkorange")
    ax.text(1 * np.pi / 2 + 0.05, 40 + PAD, "40%", ha="center", size=12,color="darkorange")
    ax.text(1 * np.pi / 2 + 0.05, 60 + PAD, "60%", ha="center", size=12,color="darkorange")
    ax.text(1 * np.pi / 2 + 0.05, 80 + PAD, "80%", ha="center", size=12,color="darkorange")
    ax.text(1 * np.pi / 2 + 0.05, 100 + PAD, "100%", ha="center", size=12,color="darkorange")

    ax.text(2 * np.pi / 2 + 0.04, 0 + PAD, "0%", ha="center", size=12,color="lightgreen")
    ax.text(2 * np.pi / 2 + 0.04, 20 + PAD, "20%", ha="center", size=12,color="lightgreen")
    ax.text(2 * np.pi / 2 + 0.04, 40 + PAD, "40%", ha="center", size=12,color="lightgreen")
    ax.text(2 * np.pi / 2 + 0.04, 60 + PAD, "60%", ha="center", size=12,color="lightgreen")
    ax.text(2 * np.pi / 2 + 0.04, 80 + PAD, "80%", ha="center", size=12,color="lightgreen")
    ax.text(2 * np.pi / 2 + 0.04, 100 + PAD, "100%", ha="center", size=12,color="lightgreen")

    ax.text(3 * np.pi / 2 + 0.02, 0 + PAD, "0%", ha="center", size=12,color="lightcoral")
    ax.text(3 * np.pi / 2 + 0.02, 20 + PAD, "20%", ha="center", size=12,color="lightcoral")
    ax.text(3 * np.pi / 2 + 0.02, 40 + PAD, "40%", ha="center", size=12,color="lightcoral")
    ax.text(3 * np.pi / 2 + 0.02, 60 + PAD, "60%", ha="center", size=12,color="lightcoral")
    ax.text(3 * np.pi / 2 + 0.02, 80 + PAD, "80%", ha="center", size=12,color="lightcoral")
    ax.text(3 * np.pi / 2 + 0.02, 100 + PAD, "100%", ha="center", size=12,color="lightcoral")

    offset += size + PAD

ax.text(0.5, 0.5 , 'Top1',
        horizontalalignment='center',
        verticalalignment='center',
        transform=ax.transAxes,
        fontsize=16)
plt.savefig(r"E:\论文\草地植物分类\model\Family_前20_top1_new1.jpg",dpi=600)
```
