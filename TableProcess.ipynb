```python
####将植物学名全称进行截断为简称
import os
import pandas as pd
import re
if __name__ == '__main__':
    table = r"E:\科研\草地植物分类\植物信息表.xlsx"
    df = pd.read_excel(table)
    for i in range(df.shape[0]):
        df.loc[i,"学名简写"] = re.findall(r"\w+",df.loc[i,"学名"])[0] + " " + re.findall(r"\w+",df.loc[i,"学名"])[1]
    with pd.ExcelWriter(r"E:\科研\草地植物分类\植物信息表_new.xlsx") as writer:
        df.to_excel(writer)
```

```python
##通过selenium获取物种名称
import os
from bs4 import BeautifulSoup
import requests
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
import selenium.webdriver.support.ui as ui
from selenium.webdriver.chrome.options import Options
from urllib.request import urlopen
from urllib.error import URLError
from time import sleep
import re
import pandas as pd
import time
import pickle
import random

if __name__ == '__main__':
    table = r"E:\科研\草地植物分类\植物信息表.xlsx"
    df = pd.read_excel(table)
    driver = webdriver.Chrome()
    driver.implicitly_wait(2)
    driver.maximize_window()
    ###########-----------------------模拟登陆-----------------------------
    logurl = "https://ppbc.iplant.cn/sp/41181"
    driver.get(logurl)
    for i in range(df.shape[0]):
        plantname = df.loc[i,"名称"]
        driver.find_element(By.ID,"txt_key").send_keys(plantname)
        driver.find_element(By.ID,"but_search").click()
        scitfullname = driver.find_element(By.XPATH,'//*[@id="result"]/div/div[1]/div[2]').text
        scicname = re.findall(r"\w+",scitfullname)[1:]
        str = " ".join(scicname)
        df.loc[i,"学名简写"] = str
    with pd.ExcelWriter(r"D:\ImageClassfication\植物信息表_new.xlsx") as writer:
        df.to_excel(writer)
```
